# Most Common Words

```python
pattern = re.compile("[a-zA-Z]+")
chaps = re.findall(pattern,book.lower())
len(chaps)
```

```python
dict_occurences = {}
for word in chaps:
    if word in dict_occurences.keys():
        dict_occurences[word] = dict_occurences[word]+1
    else:
        dict_occurences[word] = 1
```

```python
dict_occurences = [(value,key) for (key,value) in dict_occurences.items()]
```

```python
sorted(dict_occurences, reverse=True)
```

# Paragraphs where the word love was used

```python
pattern = re.compile("[^\n]+love[^\n]+")
findings = re.findall(pattern,book)
findings[:3]
```

```python
pattern = re.compile("([a-zA-Z ]+)\n\n")
findings = re.findall(pattern,book)
findings
```

```python
pattern = re.compile("[a-zA-Z]+")
chaps = re.findall(pattern,book.lower())
d = {}
for word in chaps:
    if word in d.keys():
        d[word] = d[word]+1
    else:
        d[word] = 1
    # try:
    #     return d[word]
    # except:
    #     return f"The word is not present in the book"
```

```python
d = [(value,key) for (key,value) in d.items()]
d_list = sorted(d,reverse = True)
```

```python
import nltk
from nltk.corpus import stopwords
eng_stopwords = stopwords.words("english")
```

```python
filtered = []
for count, word in d_list:
    if word not in eng_stopwords:
        filtered.append((word,count))
```

# Sentiment Analysis

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
import nltk
nltk.downloader.download('vader_lexicon')
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
analyzer.polarity_scores(book)
```

```python
import re
pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern,book)
```

```python
chapters = chapters[1:]
```

```python
for nr, chapter in enumerate(chapters):
    scores = analyzer.polarity_scores(chapter)
    print(nr+1,scores)
```
